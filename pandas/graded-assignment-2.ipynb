#### Suppose you are looking for an Airbnb for a few days in the Bronx neighborhood group. Below are the conditions the Airbnb must meet:
- 
Room type: 'Entire home/apt'- 

Price ran90 to 120$1- 20

Availability: 365 days of the year

How many Airbnb listings meet the given criteria?

```python
ng_mask = air_df['neighborhood_group'] == "Bronx"
room_type_mask = air_df['room_type'] == 'Entire home/apt'
price_range_mask = (air_df['price'] >= 90) & (air_df['price'] <= 120)
availability_mask = air_df['availability_365'] == 365
min_nights_mask = air_df['minimum_nights'] == 365

rooms_df = air_df[ng_mask & price_range_mask & availability_mask & room_type_mask]
rooms_df.shape
```
