```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
```

```python
ax = sns.scatterplot(data=tips_df, x="total_bill", y="tip", color="orange")
ax.grid()
```

```python
ax = sns.regplot(data=tips_df, x="total_bill", y="tip", color="orange")
ax.grid()
```
