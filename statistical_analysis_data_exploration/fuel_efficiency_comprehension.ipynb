```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
mpg_df = pd.read_csv('mpg.csv')
```

## Fuel Efficiency I

```python
mpg_df['origin'].unique()
```

## Fuel Efficiency II
Find the years when automobiles used 5 cylinders and describe how their horsepower has changed during those years.

```python
mpg_5cylinders_df = mpg_df[mpg_df['cylinders'] == 5]
```

```python
b1 = sns.barplot(data=mpg_5cylinders_df, x="model_year", y="horsepower", errorbar=('ci', False))
plt.suptitle("Horsepower over the years of automobiles using 5 cylinders.")
plt.title("Source: mpg dataset by upgrad.", fontsize=10)
plt.show()
```

## Fuel Efficiency III
What is the range of automobile weights from 1975–1979?

```python
filtered_df = mpg_df[(mpg_df['model_year'] >= 75) & (mpg_df['model_year'] <= 79)]
```

```python
filtered_df
```

```python
box1 = sns.pointplot(data=filtered_df, x='model_year', y="weight")
```

```python
ax = sns.pointplot(data=filtered_df, x="model_year", y="weight", 
estimator=np.mean, ci=None, join=False)
```

## Fuel Efficiency IV
