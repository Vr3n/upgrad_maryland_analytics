# Car manufacturer comprehension.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
car_manufacturer_df = pd.read_csv('./car_manufacturere_datset.csv')
```

### Problem - I
Which manufacturer has the second-highest car sales?

```python
car_manufacturer_df.columns
```

```python
car_manufacturer_df.columns = [x.lower() for x in car_manufacturer_df.columns]
```

```python
ax = sns.boxplot(data=car_manufacturer_df, x="sales_in_thousands", y="manufacturer")
```

```python
ax = sns.kdeplot(data=car_manufacturer_df, x="sales_in_thousands", hue="manufacturer", cumulative=True)
```
