## Exercise - II

Create a strip plot of the tip amount by party size and meal time.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
e2 = sns.stripplot(data=tips_df, x="size",
                   y="tip", hue="time", alpha=0.25)
sns.pointplot(data=tips_df, x="size", y="tip", hue="time", join=False, estimator=np.mean, errorbar=('ci', False))
e2.set(xlabel="Party size", ylabel="Tips")
plt.suptitle("Larger Parties Leave more in tips.",
             y=1, fontsize=16)
plt.title("Source: Seaborn Tips Dataset", fontsize=12)

handle, labels = e2.get_legend_handles_labels()

e2.legend(handle[2:], labels[2:], title="Meal", loc="upper left")

yfmt= "${:,.2f}"

ytick = mtick.FuncFormatter(lambda x, pos: yfmt.format(x))
e2.yaxis.set_major_formatter(ytick)

plt.show()
```
