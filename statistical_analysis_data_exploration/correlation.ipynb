```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
diamonds_df = sns.load_dataset("diamonds")
```

```python
corr = diamonds_df.corr(numeric_only=True)
```

```python
div_palette = sns.diverging_palette(230, 20, as_cmap=True)
ax = sns.heatmap(data=corr, annot=True, linewidths=1, cmap=div_palette, center=0, vmax=1, vmin=-1)
```
