## Problem - I
Around which year does the contribution of liquid fuels to CO2 emissions exceed that of solid fuels?

```python
sns.lineplot(data=emmision_df, x='year', y='solids', markers=True)
sns.lineplot(data=emmision_df, x='year', y='liquids', markers=True)
```

###  Problem - II
Which factor contributed the most to fossil-fuel-based CO2 emissions between 1900 and 1950?

```python
co2_df = emmision_df.set_index('year')
co2_filtered_df = co2_df[(co2_df.index >= 1900) & (co2_df.index <= 1950)]
```

```python
li = sns.lineplot(data=co2_filtered_df.drop(columns=['total', 'capita']))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
```

```python
l1 = sns.lineplot(data=co2_df, x="year", y="solids")
sns.lineplot(data=co2_df, x="year", y="liquids")
sns.lineplot(data=co2_df, x="year", y="gas")
sns.lineplot(data=co2_df, x="year", y="production")
sns.lineplot(data=co2_df, x="year", y="flaring")
```
