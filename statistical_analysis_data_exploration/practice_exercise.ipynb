## Practice with Tips
Create a side-by-side bar chart of the total revenue by day and meal.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
```

```python
tips_df = sns.load_dataset('tips')
```

```python
tips_df.head()
```

```python
e1 = sns.barplot(data=tips_df, x="day", y="total_bill",
                 hue="time", estimator=np.sum,
                 palette="Set1", errorbar=('ci', False))
# Set the axis labels.
e1.set(xlabel="Day of the week", ylabel="Total Revenue")

# Add titles.
plt.suptitle("Weekend dinners generate the most revenue.", y=1.05, fontsize=16)
plt.title("Source: Seaborn Tips data.")

# Change the legend.
plt.legend(title="Meal", loc="upper left", frameon=False)

# Change the format for the tick marks.
yfmt = '${:,.0f}'
ytick = mtick.FuncFormatter(lambda x, pos: yfmt.format(x))
e1.yaxis.set_major_formatter(ytick)

plt.show()
```
