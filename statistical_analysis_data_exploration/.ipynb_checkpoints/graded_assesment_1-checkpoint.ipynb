## Problem - I
You want to start your portfolio of Qtm coins. For that, you want to understand the price variation of these coins and know which other coin affects their price the most. Select the coin that affects the price of Qtm coins the most.

```python
crypto_corr = crypto_df.corr(numeric_only=True)
```

```python
sns.heatmap(data=crypto_corr, cmap="Blues", annot=True)
```
