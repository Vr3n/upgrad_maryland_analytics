```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
tips_df = sns.load_dataset("tips")
```

```python
ax = sns.kdeplot(data=tips_df, x="total_bill", hue="time", cumulative=True)
ax.grid()
```

```python
diamonds_df = sns.load_dataset("diamonds")
```

```python
ax = sns.kdeplot(data=diamonds_df, x="price", hue="cut")
```
