```python
from IPython.display import display, HTML

def display_dataframe_scrollable(df):
    """
    Display a Pandas DataFrame as a scrollable HTML table in JupyterLab.
    
    Parameters:
    - df: Pandas DataFrame
    
    Returns:
    - None
    """
    display(HTML(df.to_html(index=False)))
```

```python
display_dataframe_scrollable(df)
```
