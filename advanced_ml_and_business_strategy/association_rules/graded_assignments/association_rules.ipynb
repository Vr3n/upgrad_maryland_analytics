```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.DataFrame({
    'ID': [1, 2, 3, 4, 5],
    'Pens': [1, 0, 0, 1, 1],
    'Crayons': [1, 0, 1, 0, 1],
    'Paper Sheets': [1, 0, 1, 1, 1],
    'Notebooks': [0, 1, 1, 0, 0],
    'Erasers': [0, 1, 0, 0, 0]
}).set_index("ID")
```

```python
df = df.astype(bool)
```

```python
itemsets = apriori(df, min_support=0.01, use_colnames=True)
```

```python
?association_rules
```

```python
associations = association_rules(itemsets, min_threshold=0.01)
```

```python
associations.sort_values(by=['lift'], ascending=False)
```
