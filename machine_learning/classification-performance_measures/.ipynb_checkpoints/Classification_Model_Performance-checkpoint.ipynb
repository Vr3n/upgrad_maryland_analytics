# Part 1: Introduction and Data Preprocessing

```python
# Load the data
import pandas as pd
credit_df = pd.read_csv("MyCreditData.csv")
credit_df.head()
```

```python
# Ensure Python reads categorical variables as categorical
non_categorical_columns = ['duration', 'amount', 'age', 'profit']

for column in credit_df.columns:
    if column not in non_categorical_columns:
        credit_df[column] = pd.Categorical(credit_df[column])

# Create binary dependent variable 'is_profitable' using the 'profit' feature for the purpose of classification
import numpy as np
credit_df["is_profitable"] = np.where(credit_df['profit'] > 0, 1, 0)
```

```python
# Split data into input variables and target variable
y = credit_df['is_profitable']
X = credit_df.iloc[:, :-2] # All columns but the last two, profit and is_profitable

# Use dummy variables for categorical variables
X = pd.get_dummies(X, drop_first = True)

# Split the data into training and validation sets
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

# Part 2: Logistic Regression Model - Performance Evaluation
#### The best logistic regression model for this data happens to be from 'sklearn' and uses no regularization.

```python
# Build the logistic regression model for the data
from sklearn.linear_model import LogisticRegression
log_reg_model = LogisticRegression(penalty = 'none', solver = 'lbfgs', max_iter = 400).fit(X_train, y_train)
```

## Confusion Matrix

```python
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 14}) # To make the plot labels easier to read

# Plot the confusion matrix for the logistic regression model
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(log_reg_model, X_val, y_val, cmap = plt.cm.Blues);
```

## Relevant Performance Measures for Binary Classification

```python
#import the different error mesasures
from sklearn.metrics import f1_score, precision_score, recall_score, accuracy_score, roc_auc_score, roc_curve

# Predictions at threshold = 0.5
y_pred = log_reg_model.predict(X_val)

print('Validation set Accuracy: %.3f' % accuracy_score(y_val, y_pred))
print('Validation set F1 Score: %.3f' % f1_score(y_val, y_pred))
print('Validation set Sensitivity (recall of positive class): %.3f' % recall_score(y_val, y_pred, pos_label = 1))
print('Validation set Specificty (recall of negative class): %.3f' % recall_score(y_val, y_pred, pos_label = 0))
```

## ROC Curve
#### The ROC curve highlights the sensitivity - specificity trade-off

```python
# Predict probabilities for the training set
lr_probs_train = log_reg_model.predict_proba(X_train)

# Retain probabilities for only positive outcomes
lr_probs_train = lr_probs_train[:, 1]

# Predict probabilities for the validation set and retain them for only positive outcomes
lr_probs_val = log_reg_model.predict_proba(X_val)[:, 1]

# Calculate ROC AUC scores for training and validation predictions
lr_auc_train = roc_auc_score(y_train, lr_probs_train)
lr_auc_val = roc_auc_score(y_val, lr_probs_val)

# Print scores
print('ROC AUC train = %.3f' % (lr_auc_train))
print('ROC AUC val = %.3f' % (lr_auc_val))

# Calculate 1-specificity and sensitivity values for training and validation predictions
lr_1sp_train, lr_sen_train, _ = roc_curve(y_train, lr_probs_train)
lr_1sp_val, lr_sen_val, _ = roc_curve(y_val, lr_probs_val)

# Plot the ROC curve for the model
plt.plot(lr_1sp_train, lr_sen_train, marker = '.', label = 'Logistic_train')
plt.plot(lr_1sp_val, lr_sen_val, marker = '.', label = 'Logistic_val')
plt.plot([0,1], [0,1], linestyle = '--') # Baseline

# Set axis labels
plt.xlabel('1-specificity')
plt.ylabel('sensitivity')

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

## Precision-Recall Curve
#### The curve highlights the trade-off between extending credit to profitable customers and our accuracy in credit extensions

```python
# Calculate precision and recall for the validation predictions
from sklearn.metrics import precision_recall_curve
precision, recall, thresholds = precision_recall_curve(y_val, lr_probs_val)

# Create the precision-recall curve
plt.plot(recall,precision)

# Add axis labels to the plot
plt.title('Precision-Recall Curve')
plt.ylabel('Precision')
plt.xlabel('Recall')

# Show the plot
plt.show()
```

# Part 3: Misclassification Costs
#### Our first step is to understand the current profitability of the credit program, and then to is to estimate the impact of our model.

We are going to use misclassification costs to study the impact. 

We are going to use as an approximation for the false positive cost, the average profit of a negative observation, and for the false negative cost the average profit of a postive observation, based on the training set.

```python
# Store the 'profit' values in a separate array
profit_arr = np.array(credit_df['profit'])

# Split the 'profit' array into training and validation sets
profit_train, profit_val = train_test_split(profit_arr, test_size = 0.3, random_state = 1)

# Split the training set into false positives and false negatives, and calculate their means
fp_cost = abs(np.mean(profit_train[profit_train < 0]))
fn_cost = abs(np.mean(profit_train[profit_train >= 0]))

# Print the FP and FN costs
print('False Positive Cost: %.3f' % fp_cost)
print('False Negative Cost: %.3f' % fn_cost)
```

#### Note: We are interested in finding the best cut-off that will maximize the benefit of our machine learning model. We could maximize profit, in which case we would simply sum up the profit of everyone that we extend credit to. But we chose to analyze the case based on misclassification costs, but notably (perhaps as expected) it leads to the same optimal cut-off. 

#### We now calculate the misclassification cost of extending credit to everyone (in the validation set).

```python
print('"All Credit" Misclassification Cost: %.2f' % (np.count_nonzero(y_val == 0) * fp_cost))
```

#### Then, using the confusion matrix from before, we calculate the misclassification cost of extending credit just to those we predict as profitable, accounting for both false negatives and positives.

We see that our misclassification cost improves (decreases)

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_val, log_reg_model.predict(X_val)) # Matrix form of confusion matrix

fp_count = cf[0,1]
fn_count = cf[1,0]

misclassification_cost = fp_count * fp_cost + fn_count * fn_cost

print('Number of False Positives: %d' % fp_count)
print('Number of False Negatives: %d' % fn_count)
print('Prediction Misclassification Cost: %.2f' % misclassification_cost)
```

#### We now calculate the misclassification cost as we raise the cut-off value from 0 to 1 (similar to the ROC curve).

```python
# Calculate and store the misclassification costs for different values of cut-off probability
cost_train = []
cost_val=[]
for cutoff in np.arange(0, 1, 0.01):
    curr_preds = np.where(lr_probs_train > cutoff, 1, 0)
    curr_cf = confusion_matrix(y_train, curr_preds)
    curr_fp_count = curr_cf[0,1]
    curr_fn_count = curr_cf[1,0]
    curr_misclassification_cost = curr_fp_count * fp_cost + curr_fn_count * fn_cost
    cost_train.append(curr_misclassification_cost)
    curr_preds = np.where(lr_probs_val > cutoff, 1, 0)
    curr_cf = confusion_matrix(y_val, curr_preds)
    curr_fp_count = curr_cf[0,1]
    curr_fn_count = curr_cf[1,0]
    curr_misclassification_cost = curr_fp_count * fp_cost + curr_fn_count * fn_cost
    cost_val.append(curr_misclassification_cost)

# Get the X values (cut-off values)
cutoffs = np.arange(0, 1, 0.01)

# Plot misclassification cost against cut-off value
plt.plot(cutoffs,cost_train, label='Training')
plt.plot(cutoffs,cost_val, label='Validaiton')
plt.xlabel('Cut-off')
plt.ylabel('Misclassification Cost')
plt.legend()
plt.show()

# Find the minimum misclassification cost and its associated cut-off value based on the training data
best_cost = min(cost_train)
best_cutoff = cutoffs[cost_train.index(best_cost)]

#apply the cut-off value to the validation data
best_valcost=cost_val[cost_train.index(best_cost)]


print('Best Misclassification Cost on the training is %.2f at Cut-off %.3f' % (best_cost, best_cutoff));
print('Applying that cut-off to the validation data results in Misclassification Cost of %.2f ' % best_valcost);
```

#### As expected, misclassification cost is the worst when the cutoff approaches 0 or 1. Most importantly, we note we were able to improve the misclassification cost a little bit by optimizing it!
