```python
import pandas as pd
import numpy as np


df = pd.read_csv('/content/drive/MyDrive/Inference/Test.csv',header=0)
values = df.values
sec_col = values[:,1]
print(np.mean(sec_col))
```

```python
# Show distribution of variable 'Time'
import matplotlib.pyplot as plt
plt.hist(sec_col,bins=50,)
plt.xlabel("Time")
plt.show()
```

# Central Limit Theorem Demos

```python
np.random.seed(234) # so that every run yields the same result

# Here I use 10000 samples of size 2
x2 = [np.mean(np.random.choice(sec_col,size=2,replace=True,)) for i in range(10000)] # x2 is an array of sample means

plt.hist(x2,bins=50)
plt.title("n = 2")
plt.xlabel("Sample mean")
plt.show()
```
